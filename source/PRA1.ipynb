Se ha creado una función que, dada una marca y un modelo de teléfono, devuelve un dataframe con los datos de los anuncios en mercadolibre sobre ese modelo de teléfono. Por ejemplo (se buscarán solo los primeros 2 resultados):

```python
extraer_dusqueda("samsung", "z-fold4", max=2)
```

Para generar el conjunto de datos, tomaremos las marcas y modelos disponibles en la tienda oficial de Xiaomi Ecuador. De esta última página extremos la información de todo los productos existente.

```python
df_xiaomi = extraer_tienda_xiaomi()
df_xiaomi.head()
```

Renombramos el valor de Teléfonos Mi por Teléfonos Xiaomi.

```python
# Cambio la categoría Teléfono Mi por Teléfono Xiaomi
df_xiaomi.loc[df_xiaomi['marca']=="mi", 'marca'] = "xiaomi"
df_xiaomi.tail()
```

Utilizamos esta lista para buscar los anuncios de mercadolibre y extraer la información de los mismos.

```python
df = pd.DataFrame()
for k in range(len(df_xiaomi)):
    # Extraigo
    df1 = extraer_dusqueda(df_xiaomi["marca"][k], df_xiaomi["modelo"][k])
    # Agrego la información oficial
    df1["Modelo"] = df_xiaomi["modelo"][k]
    df1["Marca"] = df_xiaomi["marca"][k]
    df1["Precio oficial"] = df_xiaomi['precio'][k]
    # Agrego la información
    df = pd.concat([df, df1])

# redefino el índice
df = df.reset_index(drop=True)
df
```

Finalmente, guardamos el dataframe en un archivo csv.

```python
# Guardo el dataframe en un csv con los nombres de las columnas
df.to_csv("telefonos_xiaomi_ec.csv", index=False, header=True)
```
